```
from sklearn import datasets
from sklearn.model_selection import train_test_split

# Cargar el conjunto de datos de cáncer de mama
dataset = datasets.load_breast_cancer()
```

```
x=dataset.data
y=dataset.target
```

```
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
print(x_train.shape)
```

```
from sklearn.preprocessing import StandardScaler
escalar = StandardScaler()#CREO UN OBJETO

x_train = escalar.fit_transform(x_train)#PARA ESCALAR LOS VALORES DE X Y LO GUARDO EN X TRAIN
x_test = escalar.transform(x_test)#PARA ESCALAR LOS VALORES DE X Y LO GUARDO EN X TEST
```

```
from sklearn.linear_model import LogisticRegression#creo el modelo
algoritmo = LogisticRegression()#CREO EL OBJETO


regresion_logistica = algoritmo.fit(x_train, y_train)

y_pred = regresion_logistica.predict(x_test)
```

```
from sklearn.metrics import confusion_matrix

# 6. Generar la matriz de confusión
matriz_confusion = confusion_matrix(y_test, y_pred)

# 7. Imprimir la matriz de confusión
print("Matriz de Confusión:")
print(matriz_confusion)
```

```
from sklearn.metrics import accuracy_score

# 8. Calcular la precisión
precision = accuracy_score(y_test, y_pred)
print("Precisión:")
print(precision)
```

```
from sklearn.metrics import recall_score
#calculo la sensibilidad del modelo
# si le cambiamos el dataset que tanto cambiaria su aprendizaje 
# se denomina recall
# Calcular la sensibilidad
sensibilidad = recall_score(y_test, y_pred)
print("Sensibilidad:")
print(sensibilidad)
```

```
from sklearn.metrics import f1_score

#CALCULO PUNTAJE F1
#ERA UNA MEZCLA ENTRE PRECISION Y EXACTITUD
# Calcular el puntaje F1
puntaje_f1 = f1_score(y_test, y_pred)
print("Puntaje F1:")
print(puntaje_f1)
```

```
from sklearn.metrics import roc_auc_score

# Calcular el puntaje AUC
puntaje_auc = roc_auc_score(y_test, y_pred)
print("Puntaje AUC:")
print(puntaje_auc)
```

```
#CALCULO LA CURVA ROC-AUC

from sklearn.metrics import roc_auc_score

roc_auc = roc_auc_score (y_test, y_pred)

print ('CURVA DE ROC AUC: \n', roc_auc)



#UMBRAL DE ROC-AUC

#SENSIBILIDAD = VP/ (VP + FN), TASA DE VP
#ESPECIFICIDAD = VN / (VN + FP), TASA DE VN
#SUBE SENSIBILIDAD Y AUMENTA ESPECIFICIDAD
#BAJO SENSIBILIDAD Y SUBE ESPECIFICIDAD
# SENSIBILIDAD (TASA DE VP) = 1-ESPECIFICIDAD (TASA DE FALSOS POSITIVOS)
```

```
from sklearn import datasets

dataset = datasets.load_breast_cancer()



X= dataset.data
y= dataset.target

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split (X, y, test_size=0.2, random_state=11)


# se escalan los datos

from sklearn.preprocessing import   StandardScaler

#instancio creando el objeto escalar
escalar = StandardScaler ()

X_train = escalar.fit_transform (X_train)
X_test = escalar.fit_transform (X_test)

#defino el algoritmo a utilizar

from sklearn.linear_model import LogisticRegression

algoritmo = LogisticRegression()

#entreno el modelo

algoritmo.fit(X_train, y_train)

#realizo la predicción

y_pred = algoritmo.predict(X_test)


#verifico matriz de confusion

from sklearn.metrics import confusion_matrix

matriz = confusion_matrix (y_test, y_pred)

print ('Matriz de confusion: \n', matriz)

# calculo la precision del modelo

from sklearn.metrics import precision_score

precision = precision_score (y_test, y_pred)

print ('Precision: \n', precision)

#calculo la exactidad del modelo

from sklearn.metrics import accuracy_score

exactitud = accuracy_score (y_test, y_pred)

print ('Exactitud: \n', exactitud)

#calculo la sensibilidad del modelo
# si le cambiamos el dataset que tanto cambiaria su aprendizaje 
# se denomina recall

from sklearn.metrics import recall_score

sensibilidad = recall_score (y_test, y_pred)

print ('Sensibilidad: \n', sensibilidad)

#CALCULO PUNTAJE F1
#ERA UNA MEZCLA ENTRE PRECISION Y EXACTITUD

from sklearn.metrics import f1_score

puntaje_f1 = f1_score (y_test, y_pred)

print ('Puntaje F1: \n', puntaje_f1)

#CALCULO LA CURVA ROC-AUC

from sklearn.metrics import roc_auc_score

roc_auc = roc_auc_score (y_test, y_pred)

print ('CURVA DE ROC AUC: \n', roc_auc)



#UMBRAL DE ROC-AUC

#SENSIBILIDAD = VP/ (VP + FN), TASA DE VP
#ESPECIFICIDAD = VN / (VN + FP), TASA DE VN
#SUBE SENSIBILIDAD Y AUMENTA ESPECIFICIDAD
#BAJO SENSIBILIDAD Y SUBE ESPECIFICIDAD
# SENSIBILIDAD (TASA DE VP) = 1-ESPECIFICIDAD (TASA DE FALSOS POSITIVOS)
```
