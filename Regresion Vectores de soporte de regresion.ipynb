```

#Los kernels son funciones que transforman los datos en un espacio de mayor dimensión donde se pueden separar más fácilmente. Los tipos comunes de kernels incluyen:
#linear: Kernel lineal.
#poly: Kernel polinómico.
#rbf: Función de base radial (kernel gaussiano).
#sigmoid: Kernel sigmoide.

C=1.0:

C es el parámetro de regularización. Controla la compensación entre lograr un margen amplio y minimizar el error de clasificación en el conjunto de entrenamiento.
Un valor más alto de C da mayor importancia a ajustar perfectamente todos los puntos de entrenamiento, permitiendo menos errores pero posiblemente reduciendo la capacidad de generalización.
Un valor más bajo de C permite un margen más amplio, permitiendo algunos errores en los puntos de entrenamiento pero mejorando la capacidad de generalización.

C=1.0:

C es el parámetro de regularización. Controla la compensación entre lograr un margen amplio y minimizar el error de clasificación en el conjunto de entrenamiento.
Un valor más alto de C da mayor importancia a ajustar perfectamente todos los puntos de entrenamiento, permitiendo menos errores pero posiblemente reduciendo la capacidad de generalización.
Un valor más bajo de C permite un margen más amplio, permitiendo algunos errores en los puntos de entrenamiento pero mejorando la capacidad de generalización.

Parámetros Adicionales del SVR
degree:

Este parámetro es relevante solo si utilizas un kernel polinómico (poly).
Especifica el grado del polinomio.
Ejemplo: SVR(kernel='poly', degree=3, C=1.0, epsilon=0.2)
gamma:

Es relevante para los kernels rbf, poly y sigmoid.
Define cómo influye cada punto de datos en el cálculo del modelo. Un valor alto significa que cada punto de datos tiene una influencia estrecha (menos suavización), mientras que un valor bajo significa que cada punto de datos tiene una influencia amplia (más suavización).
Puede ser un valor flotante o 'scale' (1 / (n_features * X.var())) o 'auto' (1 / n_features).
Ejemplo: SVR(kernel='rbf', gamma=0.1, C=1.0, epsilon=0.2)
coef0:

Es relevante para los kernels poly y sigmoid.
Especifica el término independiente en el kernel.
Ejemplo: SVR(kernel='poly', degree=3, coef0=1, C=1.0, epsilon=0.2)
shrinking:

Un booleano (True o False) que especifica si se utiliza la heurística de reducción del problema.
La heurística puede hacer que el algoritmo sea más rápido en algunos casos.
Ejemplo: SVR(kernel='linear', C=1.0, epsilon=0.2, shrinking=True)
tol:

Tolerancia para el criterio de parada.
Especifica el umbral de convergencia. Un valor más bajo puede resultar en un modelo más preciso, pero puede requerir más tiempo de computación.
Ejemplo: SVR(kernel='linear', C=1.0, epsilon=0.2, tol=1e-4)
max_iter:

Número máximo de iteraciones. Un valor de -1 significa que no hay límite en el número de iteraciones.
Controla cuánto tiempo debe ejecutar el algoritmo antes de detenerse.
Ejemplo: SVR(kernel='linear', C=1.0, epsilon=0.2, max_iter=1000)

sugerido
Para explorar un modelo menos sensible a los puntos atípicos y con mayor capacidad de generalización:

C = 1.0
epsilon = 0.1
Para un modelo que ajuste más a los datos de entrenamiento:

C = 10.0
epsilon = 0.05
Para un modelo con un margen de tolerancia más amplio y menos ajustado a los datos de entrenamiento:

C = 0.1
epsilon = 0.2
```

```
########## LIBRERÍAS A UTILIZAR ########## 
#Se importan la librerias a utilizar 
import numpy as np 
import matplotlib.pyplot as plt 
from sklearn import datasets 
() 
### #### PREPARAR LA DATA ####  
#Importamos los datos de la misma librería de scikit-learn 
boston=datasets. load_boston() 
print(boston) 
print() 
########## ENTENDIMIENTO DE LA DATA ########## 
#Verifico la información contenida en el dataset 
print('Información en el dataset:') 
print(boston.keys()) 
print() 
()
#Verifico las características del dataset 
print('Características del dataset:') 
print (boston.DESCR) 
()
#Verifico la cantidad de datos que hay en los dataset 
print('Cantidad de datos:') 
print (boston.data.shape) 
print() 
#Verifico la información de las columnas 
print('Nombres columnas :') 
print (boston.feature_names) 
()
#PREPARAR LA DATA VECTORES DE SOPORTE REGRESIÓN 
# Seleccionamos solamente la columna 6 del dataset 
X_svr=boston.data[:, np.newaxis, 5] 
()
#Defino los datos correspondientes a las etiquetas 
y_svr=boston.target 
()
#Graficamos los datos correspondientes 
plt.scatter (X_svr, y_svr) 
plt.show() 
()
########## IMPLEMENTACIÓN DE VECTORES DE SOPORTE REGRESIÓN ########## 
from sklearn.model_selection import train_test_split 
()
#Separo los datos de "train" en entrenamiento y prueba para probar los algoritmos 
X_train, X_test, y_train, y_test = train_test_split(X_svr, y_svr, test_size=0.2)
from sklearn.svm import SVR 
()
#Defino el algoritmo a utilizar      define el modelo de Máquina de Vectores de Soporte para Regresión (Support Vector Regression, SVR
svr = SVR(kernel='linear', C=1.0, epsilon=0.2) 
()
#Entreno el modelo 
svr.fit(X_train, y_train) 
()
#Realizo una predicción 
Y_pred = svr.predict(X_test) 
()
#Graficamos los datos junto con el modelo 
plt.scatter (X_test, y_test) 
plt.plot(X_test, Y_pred, color='red', linewidth=3) 
plt.show() 
()
print() 
print('DATOS DEL MODELO VECTORES DE SOPORTE REGRESIÓN') 
print() 
print('Precisión del modelo:') 
print(svr.score (X_train, y_train)) 
```
