# Ejemplo de Regresion Lineal

```
import numpy as np
import matplotlib.pyplot as plt
()
X = [11,16,8,12]
y=[6,7,4,5]
()
X_ones = np.ones(4)
X_ones
()
X_ones = np.c_[X_ones, X]#CONCATENAMOS UN VECTOR CON UNOS CON EL VECTOR X
X_ones
()
coefFijos = np.linalg.inv(X_ones.T.dot(X_ones)).dot(X_ones.T).dot(y)
coefFijos
()
plt.axis([7,20,2,9])
plt.xlabel('Cantidad de Horas')
plt.ylabel('Calificacion')
plt.title('Nivel de Aprobacion x Horas de Estudio')
plt.grid()
plt.scatter(X,y,s=40, c='#06d6a0')
X_lim = [7, 20]
X_lim_ones=np.c_[np.ones(2), X_lim]
y_lim = X_lim_ones.dot(coefFijos)
plt.plot(X_lim, y_lim, 'r-')
()
califOctav = coefFijos[0]+coefFijos[1]*10
califOctav

```

# Regresion lineal con calculo del coeficiente de determinacion

```
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score

# Datos de entrada
X = [11, 16, 8, 12]
y = [6, 7, 4, 5]

# Agregar una columna de unos para el término independiente
X_ones = np.c_[np.ones(4), X]

# Calcular los coeficientes de la regresión lineal
coeficientes = np.linalg.inv(X_ones.T.dot(X_ones)).dot(X_ones.T).dot(y)
print('Coeficientes de la regresión lineal:', coeficientes)

# Graficar los puntos y la línea de regresión
plt.scatter(X, y, s=40, c='#06d6a0')
plt.xlabel('Cantidad de Horas')
plt.ylabel('Calificación')
plt.title('Nivel de Aprobación vs Horas de Estudio')
plt.grid()

# Calcular la línea de regresión
X_min = min(X)
X_max = max(X)
X_line = np.linspace(X_min, X_max, 100)
X_line_ones = np.c_[np.ones(100), X_line]
y_line = X_line_ones.dot(coeficientes)

# Graficar la línea de regresión
plt.plot(X_line, y_line, 'r-', label='Regresión Lineal')

# Calcular el coeficiente de determinación (R^2)
y_pred = X_ones.dot(coeficientes)
r2 = r2_score(y, y_pred)
print('Coeficiente de determinación (R^2):', r2)

# Calcular la calificación esperada para 10 horas de estudio
horas_estudio = 10
calificacion_esperada = coeficientes[0] + coeficientes[1] * horas_estudio
print(f'Calificación esperada para {horas_estudio} horas de estudio:', calificacion_esperada)

# Mostrar leyenda y la gráfica
plt.legend()
plt.show()
```
