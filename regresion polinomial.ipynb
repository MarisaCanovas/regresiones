Cómo Detectar y Mitigar el Overfitting
Para mitigar el overfitting en la regresión polinomial, es importante utilizar técnicas como:

Validación Cruzada: Utilizar validación cruzada (como K-fold cross-validation) para evaluar el modelo en múltiples particiones de los datos. Esto ayuda a verificar si el modelo generaliza bien a datos no vistos y no solo a los datos de entrenamiento.

Regularización: Aplicar técnicas de regularización como la regresión Ridge (L2) o la regresión Lasso (L1) para penalizar coeficientes grandes. Esto ayuda a reducir la complejidad del modelo y a prevenir el overfitting.

Selección de Características: Realizar una selección cuidadosa de características o utilizar métodos automatizados para seleccionar características relevantes y eliminar características irrelevantes que puedan estar contribuyendo al overfitting.

Monitorización del Rendimiento: Evaluar continuamente el rendimiento del modelo en datos de prueba y ajustar el grado del polinomio según sea necesario para encontrar un equilibrio entre ajuste y generalización.

# Validación Cruzada k-Fold

El método Holdout (train/test split) es una técnica básica pero efectiva para evaluar modelos de aprendizaje automático. Aunque tiene algunas limitaciones, es un buen punto de partida y es particularmente útil cuando se necesita una evaluación rápida. Para una evaluación más robusta y fiable, es recomendable complementarlo con técnicas como la validación cruzada k-Fold.  La validación cruzada k-Fold divide el conjunto de datos en k partes (folds) y usa cada parte como conjunto de prueba en una iteración, mientras las partes restantes se usan como conjunto de entrenamiento. Esto se repite k veces, y se promedian los resultados para obtener una estimación robusta del rendimiento del modelo.

```
# Importamos las librerías necesarias
from sklearn import datasets, linear_model
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split, KFold, cross_val_score
from sklearn.preprocessing import PolynomialFeatures

# Importamos los datos de Boston
boston = datasets.load_boston()

# Seleccionamos solamente la columna 5 del dataset (RM: promedio de habitaciones por vivienda)
X_p = boston.data[:, np.newaxis, 5]
y_p = boston.target

# Graficamos los datos
plt.scatter(X_p, y_p)
plt.title('Datos Originales')
plt.xlabel('Número Promedio de Habitaciones (RM)')
plt.ylabel('Valor Mediano de las Viviendas ($1000s)')
plt.show()

# Dividimos los datos en conjunto de entrenamiento y prueba
X_train_p, X_test_p, y_train_p, y_test_p = train_test_split(X_p, y_p, test_size=0.2, random_state=42)

# Definimos el grado del polinomio
degree = 2

# Creamos el objeto PolynomialFeatures
poly_features = PolynomialFeatures(degree=degree)

# Transformamos las características a polinomiales
X_train_poly = poly_features.fit_transform(X_train_p)
X_test_poly = poly_features.transform(X_test_p)

# Creamos el modelo de regresión lineal
model = linear_model.LinearRegression()

# Aplicamos validación cruzada con k=5 folds
kf = KFold(n_splits=5, shuffle=True, random_state=42)# spli:5 subdivisiones, shufle es ordenamiento aleatorioy random es la semilla

# Evaluamos el modelo usando cross_val_score
scores = cross_val_score(model, X_train_poly, y_train_p, cv=kf, scoring='r2')

# Entrenamos el modelo
model.fit(X_train_poly, y_train_p)

# Realizamos predicciones
y_pred = model.predict(X_test_poly)

# Graficamos los datos y la línea de regresión polinomial
plt.scatter(X_test_p, y_test_p, color='blue', label='Datos reales')
plt.plot(X_test_p, y_pred, color='red', linewidth=3, label='Regresión Polinomial (grado 2)')
plt.title('Regresión Polinomial sobre los Datos de Test')
plt.xlabel('Número Promedio de Habitaciones (RM)')
plt.ylabel('Valor Mediano de las Viviendas ($1000s)')
plt.legend()
plt.show()

# Mostramos métricas de rendimiento del modelo
print('Coeficientes del modelo de regresión polinomial:')
print(model.coef_)
print('Intercepto del modelo de regresión polinomial:')
print(model.intercept_)
print('Precisión del modelo en datos de entrenamiento:')
print(model.score(X_train_poly, y_train_p))
```

Intercepto en el Contexto de Validación Cruzada
En el proceso de validación cruzada, dividimos los datos en múltiples particiones (folds) y utilizamos estas particiones para entrenar y evaluar el modelo de manera repetida. Durante este proceso, se obtienen múltiples modelos y, por lo tanto, múltiples valores de intercepto.

Para calcular y entender el intercepto en el contexto de la validación cruzada, aquí hay algunos puntos clave:

Cálculo del Intercepto: En cada iteración de la validación cruzada (por ejemplo, en cada fold), se entrena un modelo de regresión polinomial que incluye el cálculo del intercepto 𝛽0

Promedio de Interceptos: Después de completar todas las iteraciones de validación cruzada, se puede promediar el valor del intercepto 𝛽0 de todos los modelos obtenidos en cada fold. Esto proporciona una estimación más robusta del intercepto del modelo de regresión polinomial que considera la variabilidad entre diferentes folds.

Utilización del Intercepto: El intercepto promedio obtenido se utiliza para evaluar y comparar el rendimiento del modelo en diferentes particiones de datos. Es una medida importante porque indica el valor esperado de la variable dependiente cuando todas las características son cero, lo cual es relevante para interpretar las predicciones del modelo.
